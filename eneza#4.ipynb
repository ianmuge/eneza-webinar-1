# Preliminary Report
- There is no visible correlation between the popularity of majors and median salary recorded. On the record, there are records of highest median is on a major that has a very small sample size. The record wiith the highest sample size is averaging at about 40000 which is in the middle of the median salary sample
- There is a very loose negative correlation between the median salary and the share that was women of that particular major. There are notable outliers with a number records for example a record that intersects at 0.1 share and 110000 median and at 40000 salary and 0.0 share
- There is no direct correlation between the number of full-time employees and median salary same case as the popularity of majors and median comparison
